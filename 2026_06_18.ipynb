```python
#8. 1부터 5까지의 합을 출력해주세요.

i = 1 #시작값
limit = 5 # 종료값
sum = 0 #더할 값을 저장할 변수, 초기값 0으로 설정

#i값이 limit(5)까지가 참이며 i값이 limit + 1(6)이면 거짓으로 종료
while i <= limit:
  """
  sum값이 초기 0부터 while문이 반복될 때 마다 i값을 더해 줌
  증가되는 순서와 합해지는 값을 주의
  "sum = sum + i" 구문이 "i = i + 1" 밑에 있으면 첫 번째 반복문일 때
  i가 1이 더해진 2부터 시작되고, 마지막 반복문일 때 i는 6이 되어
  2부터 6까지의 합을 구하게 됨
  """
  sum = sum + i
  #print(i, sum)
  #i값이 초기 1부터 반복될 때 마다 1씩 증가됨
  i = i + 1

#while 밖에서 출력해야 최종으로 더한 값이 출력
#while 안에 있으면 매번 더한값이 출력됨
print(sum)
```

```python
#9. 1부터 100까지의 수 중에서 짝수의 합을 출력해주세요. while, if, %

i = 1 #시작값
limit = 100 #종료값
sum = 0 #더할 값을 저장할 변수, 초기값 0으로 설정

#i값이 limit(100)까지가 참이며 i값이 limit + 1(101)이면 거짓으로 종료
while i <= limit:
  #i값을 2로 나누었을 때 나머지가 0, 즉 짝수인지 확인
  #참이면 i값을 출력, 거짓이면 다시 while문 반복
  if i % 2 == 0:
    #sum값이 초기 0부터 while문이 반복될 때 마다 i값을 더해 줌
    sum = sum + i

  """
  i값이 초기 1부터 반복될 때 마다 1씩 증가됨
  if문 안에 들어가지 않도록 조심
  if문 안에 들어가면 i가 1일 때 홀수라서 거짓이기 때문에 1이 증가하지 않아서
  while문이 무한루프에 빠지게 됨
  """
  i = i + 1

print(sum)
```

```python
#1. 1부터 숫자를 1, 2, 3... 계속 더해 나갈 때, 그 합이 처음으로 1000을 넘게 하는 숫자는 무엇인지 출력해주세요.

i = 1 #시작값
sum = 0 #더할 값을 저장할 변수, 초기값 0으로 설정
limit = 1000 #종료값

#sum값이 limit(1000)까지가 참이며, sum값이 limit + 1(1001)이면 거짓으로 종료
while sum <= limit:
  #sum값이 초기 0부터 while문이 반복될 때 마다 i값을 더해 줌
  sum += i
  #print(i, sum)
  #i값이 초기 1부터 반복될 때 마다 1씩 증가됨
  i += 1

#while문이 마지막으로 실행될 때 i값이 1증가 했으므로 최종 결과를 구할 때 1을 빼줘야 됨
print(i - 1)
```

```python
#2. 구구단 1단부터 9단까지의 모든 결과값 중에서, 홀수인 결과값들만 모두 더한 총합을 출력해주세요.

i = 1
j = 1
sum = 0 #더할 값을 저장할 변수, 초기값 0으로 설정

#j값이 9까지가 참이며 j값이 10이면 거짓으로 종료
while j <= 9:
  #i값이 9까지가 참이며 i값이 10이면 거짓으로 종료
  while i <= 9:
    #print("{} * {} = {}".format(j, i, j * i))
    #구구단의 결과를 2로 나누었을 때 몫이 0이 아닌, 즉 홀수인지 확인
    if (j * i) % 2 != 0:
      #print(j * i)
      #구구단의 결과가 홀수일 때 그 값을 계속 더함
      sum += (j * i)
    #i값이 초기 1부터 반복될 때 마다 1씩 증가됨
    i += 1
  #내부 while문이 종료되었을 때 i값이 10됨
  #다시 1부터 시작해야 되므로 i값을 1로 초기화
  i = 1
  #j값이 초기 1부터 반복될 때 마다 1씩 증가됨
  j += 1
print(sum)
```
