```python
# 클래스 : 객체를 표현하는 여러가지 기법 중 1

# 대학병원 환자 관리 시스템을 만들자!
# 환자등록, 진료 기록, 약처방 내역, 입원/ 외래...

# 왜 클래스가 필요함???
# 환자가 1명이라면?
name = "김환자"
age = 42
bp = 120

# 환자가 100명이면?
# name1,age1,bp1,name2,age2,bp2...name100,age100,bp100 -> 변수가 너무 많아지는데?

# 1. 클래스란? -> 객체를 만들기 위한 '양식(설계도)'

# 클래스 : 병원에 비치된 '빈 차트 양식' -> 1개
#   객체 : 그 양식으로 찍어낸 '실제 환자 차트' -> 환자 수 만큼

# 클래스의 구성 요소
# - 속성(attribute) : 차트에 적는 칸 -> 이름, 나이, 전화번호, 혈압 (명사), 변수인데 클래스 안쪽에 있는거
# - 기능(method) : 해당 환자한테 하는 행위 -> 처방하다, 입원하다 (동사), 함수인데 클래스 안쪽에 있는거

# 환자 클래스
# 속성 : 이름 나이, 혈압, 진단명...
# 기능 : 접수하다, 처방하다, 퇴원하다...

class Patient:
  def check_in(self):           # 메서드 1 : 접수
    print("환자 접수 완료")
  def discharge(self):          # 메서드 2 : 퇴원
    print("퇴원 처리 완료")

kim = Patient()                 # kim은 Patient 양식(클래스)로 찍어낸 실제 객체 == 인스턴스

kim.check_in()
kim.discharge()
```

```python
# 2. 객체(Object) vs 인스턴스(Instance) -> 약간 말장난인가? 싶지만 중요함
# - 대상 그 자체를 가리킬 때  -> 객체
# - "이건 ~~ 클래스로 만든 거야" 라고 클래스랑 엮어서 말하면 -> 인스턴스

a = int(10) # a 는 int 클래스의 인스턴스
print(a)

b = list(range(10)) # b 는 list 클래스의 인스턴스
b.append(10)        # append list 클래스의 메서드였었다!
print(b)

c = dict(x=10, y=20) # c 는 dict 클래스의 인스턴스
print(c)

# b.append(10)는, 위에서 만든 kim.check_in과 똑같은 원리
# '객체가 자기 기능(메서드)을 실행하는 것' ----> 이게 객체지향이 추구하는 것
```

```python
# 3. self -> 한 기능에서 다른 기능을 부르고 싶을 때
# 접수(check_in)을 하면 자동으로 차트 출력(print_chart)도 같이 하고싶어!
# 이럴 때 '이 차트 본인'을 가리키는 self 키워드를 쓴다.

class Patient:
  def print_chart(self):
    print("==환자 차트==")
  def check_in(self):
    print("접수 시작")
    self.print_chart() # self = '이 환자 본인' -> 본인의 다른 메서드 호출

kim = Patient()
kim.check_in() # 기능이 실제로 실행되려면 인스턴스가 있어야해, 접수 하면서 차트까지 같이 출력됨
```

```python
# 4. __init__ (생성자, constructor) -> '신규환자 접수' 순간에 자동으로 실행되는 메서드
# 환자를 등록(인스턴스 생성)하는 바로 그 순간 자동으로 호출되는 메서드
# 보통은 여기서 기본 정보를 적어 넣는다.

class Patient:
  def __init__(self):                              # 생성자
    print("__init__ 실행됨 -> 신규 환자 접수!")
    self.greeting = "안녕하세요, 환자분~~~"

  def greet(self):
    print(self.greeting)

kim = Patient() # 이 순간 __init__이 자동 실행됨
lee = Patient() # 환자 한 명 더 등록 -> __init__ 또 실행됨

kim.greet()

# 참고 : 일반 함수랑 비교
def plus(a, b):
  z = 10
  print(a + b)

# def b():
#   print(z)

plus(1, 2)
# b()
# 함수는 내가 직접 불러야 실행되지만
# __init__ 생성자는 인스턴스를 만드는 순간 '자동'으로 실행되는 점이 차이점이다.
```

```python
# 5. 속성 -> 환자마다 다른 정보를 차트에 담기
# 접수할 때 이름과 나이를 받아서 해당 환자의 차트에 적는다.

class Patient:
  def __init__(self, name, age): # 생성자의 매개변수가 여러개
    self.greeting = "안녕하세요"
    self.name = name             # 넘겨받은 값을 해당 변수(속성)에 저장
    self.age = age

  def introduce(self):
    # 클래스 '내부'에서 속성에 접근할 때는 'self.속성' 으로 써야한다
    print("{} 저는 {}살 {}입니다.".format(self.greeting, self.age, self.name))

kim = Patient("김환자", 56) # 이 때 생성자는 실행되기 때문에 매개변수에 맞게 인자 전달
kim.introduce()
print("이름 : {}".format(kim.name)) # 클래스 '외부'에서 접근 -> '인스턴스.속성'으로 써야한다
print("나이 : {}".format(kim.age))
```

```python
# 6. 여러 환자 -> 차트 하나에 환자 한 명씩(인스턴스가 여러 개)
# 양식(클래스)은 하나지만, 찍어낸 차트(인스턴스)는 서로 독립이다.

kim = Patient("김환자", 56) # 차트를 만들면서 이름, 나이 칸을 채운다.
kim.introduce()
print("이름 : {}".format(kim.name))

lee = Patient("이영희", 33)
lee.introduce()
print("이름 : {}".format(lee.name))

# kim의 정보와 lee의 정보는 절대로 섞이지 않는다. 각자 본인 차트가 있음
```

```python
# 7. 빈 차트 만들고 나중에 채우기
# 생성자 없이 일단 환자만 등록해두고 정보는 나중에 채울 수도 있다.

class Patient:
  pass

kim = Patient() # 빈 차트 발급
kim.name = "김철수" # 빈 차트에 나중에 이름 칸을 채우는 행위
print(kim.name)

lee = Patient()
lee.age = 33
print(lee.age)
```
