```python
#18-3. 인자로 넘겨받은 정수의 모든 약수를 출력하는 함수를 구현해주세요

#약수: 어떤수를 나누었을 때 나머지가 0
def print_divisors(num):
  #양수의 약수
  if num > 0:
    i = 1
    #양수일 경우 1부터 입력된 값까지 순차적으로 증가
    while i <= num:
      if num % i == 0:
        print(i)
      i += 1
  #음수의 약수
  elif num < 0:
    j = -1
    #음수일 경우 -1부터 입력된 값까지 순차적으로 감소
    while j >= num:
      if num % j == 0:
        print(j)
      j -= 1
  #0의 약수
  else:
    print("0의 약수는 0이 아닌 모든 정수입니다.")

print_divisors(1000)
```

```python
#20-6. 입력받은 정수의 모든 약수의 합을 리턴하는 함수를 구현해주세요.

def get_divisors_sum(num):
  s = 0 #약수의 합을 저장하는 변수
  i = 1

  while i <= num:
    if num % i == 0:
      s += i
    i += 1
  return s

print("입력받은 정수의 모든 약수의 합은 : {} 입니다".format( get_divisors_sum(100)))
```

```python
#20-7. 입력받은 정수가 소수인지 아닌지 알려주는 함수를 구현해주세요.
#소수: 1보다 큰 자연수 중에서 1과 자기 자신만을 약수로 가지는 수, 즉, 약수가 정확히 두 개인 자연수
def is_prime_number(num):
  s = 1 #약수의 합을 저장하는 변수, 1은 항상 포함되기 때문에 1로 설정
  i = 2 #1은 소수가 아니기 때문에 2부터 시작

  #num이 매우 클 경우 연산량이 많음
  while i <= num:
    if num % i == 0:
      #약수의 갯수를 저장
      s += 1
    i += 1

  #소수의 약수는 2개
  if s <= 2:
    return True
  else:
    return False

print("3은 소수입니다 : {}".format(is_prime_number(3)))
print("4는 소수입니다 : {}".format(is_prime_number(4)))
print("1000은 소수입니다 : {}".format(is_prime_number(1000)))
```
