```python
# 문제 : 할인 대상인지 아닌지 출력해주세요
# 조건 : 나이가 19세 이하이거나 60세 이상이면 할인 대상입니다
# 조건 : and, or 없이 해결하는 방법이 있어야 합니다
# 조건 : 8가지 이상의 방식으로 풀어야합니다

age = 19 # <- 이 숫자를 바꿔가면서 실행해보세요

#1-1 순차 조건문
print("== 정답 v1-1==")
if age <= 19:
  print("할인 대상입니다")
elif age >= 60:
  print("할인 대상입니다")
else:
  print("할인 대상이 아닙니다")

#1-2 순차 조건문
print("== 정답 v1-2==")
if age >= 60:
  print("할인 대상입니다")
elif age <= 19:
  print("할인 대상입니다")
else:
  print("할인 대상이 아닙니다")

#2-1 이중 조건문
print("== 정답 v2-1==")
if age > 19:
  if age < 60:
    print("할인 대상이 아닙니다")
  else:
    print("할인 대상입니다")
else:
  print("할인 대상입니다")

#2-2 이중 조건문
print("== 정답 v2-2==")
if age < 60:
  if age > 19:
    print("할인 대상이 아닙니다")
  else:
    print("할인 대상입니다")
else:
  print("할인 대상입니다")

#3 or 연산자
print("== 정답 v3==")
if age <= 19 or age >= 60:
  print("할인 대상입니다")
else:
  print("할인 대상이 아닙니다")

#4 and 연산자
print("== 정답 v4==")
if age >19 and age < 60:
  print("할인 대상이 아닙니다")
else:
  print("할인 대상입니다")

#5 not and not 연산자
print("== 정답 v5==")
if not (age <= 19) and not(age >= 60):
  print("할인 대상이 아닙니다")
else:
  print("할인 대상입니다")

#6 not or not 연산자
print("== 정답 v6==")
if not (age > 19) or not (age < 60):
  print("할인 대상입니다")
else:
  print("할인 대상이 아닙니다")

#7 not or 연산자
print("== 정답 v7==")
if not (age > 19) or age >= 60:
  print("할인 대상입니다")
else:
  print("할인 대상이 아닙니다")

#8 or not 연산자
print("== 정답 v8==")
if age <= 19 or not (age < 60):
  print("할인 대상입니다")
else:
  print("할인 대상이 아닙니다")

```
