```python
#21-1. 1부터 1000사이에 존재하는 소수들의 개수를 출력해주세요.(개념 : 함수, 반복문, 리턴)

def is_prime_number(num):
  #소수: 1보다 큰 자연수 중에서 1과 자기 자신만을 약수로 가지는 수
  #1이 입력되면 거짓 반환
  if num <= 1:
    return False

  i = 2 #1은 소수가 아니기 때문에 2부터 시작
  #소수: 1과 자기 자신만 약수, 따라서 자신과 나누는 경우는 제외
  while i < num:
    if num % i == 0:
      #약수가 확인될 경우 거짓을 반환
      return False
    #다음수로 이동
    i += 1
  #약수가 없는 경우 참을 반
  else:
    return True

sum = 0 #소수의 갯수를 저장하는 변수
j = 1

#1부터 1000까지 순차적으로 증가하며 소수인지 확인
while j <= 1000:
  #소수를 찾는 함수가 참이면 소수의 갯수를 1 증가
  if(is_prime_number(j) == True):
    sum += 1

  #다음수로 순차적 증가
  j += 1

print(sum)
```

```python
#21-3. 인자로 전달받은 숫자가 10이라고 할때 1부터 10 사이에 존재하는 모든 소수의 개수를 리턴하는 함수 구현

def count_prime_number(num):
  #소수인지 확인하는 함수
  def is_prime_number(num):
    #소수: 1보다 큰 자연수 중에서 1과 자기 자신만을 약수로 가지는 수
    #1이 입력되면 거짓 반환
    if num <= 1:
      return False

    i = 2 #1은 소수가 아니기 때문에 2부터 시작
    #소수: 1과 자기 자신만 약수, 따라서 자신과 나누는 경우는 제외
    while i < num:
      if num % i == 0:
        #약수가 확인될 경우 거짓을 반환
        return False
      #다음수로 이동
      i += 1
    #약수가 없는 경우 참을 반
    else:
      return True

  count = 0 #소수의 갯수를 저장하는 변수
  j = 1

  #2부터 num까지 순차적으로 증가하며 소수인지 확인
  while j <= num:
    #소수이면 갯수를 1 증가
    if(is_prime_number(j) == True):
      count += 1

    #다음수로 순차적 증가
    j += 1

  #저장된 소수의 갯수를 반환
  return count

count_prime_number(10)
```

```python
#21-4. 인자로 전달받은 숫자가 10이라고 할때 1부터 10 사이에 존재하는 모든 소수의 합을 리턴하는 함수 구현

def sum_prime_number(num):
  #소수인지 확인하는 함수
  def is_prime_number(num):
    #소수: 1보다 큰 자연수 중에서 1과 자기 자신만을 약수로 가지는 수
    #1이 입력되면 거짓 반환
    if num <= 1:
      return False

    i = 2 #1은 소수가 아니기 때문에 2부터 시작
    #소수: 1과 자기 자신만 약수, 따라서 자신과 나누는 경우는 제외
    while i < num:
      if num % i == 0:
        #약수가 확인될 경우 거짓을 반환
        return False
      #다음수로 이동
      i += 1
    #약수가 없는 경우 참을 반환
    else:
      return True

  sum = 0 #소수의 합을 저장하는 변수
  j = 1

  #2부터 num까지 순차적으로 증가하며 소수인지 확인
  while j <= num:
    #소수이면 합을 더함
    if(is_prime_number(j) == True):
      sum += j

    #다음수로 순차적 증가
    j += 1

  #저장된 소수의 합을 반환
  return sum

sum_prime_number(10)
```
